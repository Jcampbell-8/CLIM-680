## **Shift Longitudes**

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

### HADISST

```python
path='/shared/obs/precip/cmap/monthly/precip.mon.mean.nc'
```

```python
ds_cmap=xr.open_dataset(path)
ds_cmap
```

```python
precip_check=ds_cmap['precip'].sel(lat=0,
                                time=ds_cmap['time'][0],
                                method='nearest')
plt.plot(ds_cmap['lon'],precip_check)
plt.title('CMAP Precipitation -180-180')
```

lons=0; new_lons=360%360=0
lons=180; new_lons=540%360=180
lons=-100; new_lons=260%360=100
lons=100; new_lons=460%360=260

```python
new_lons=((ds_cmap['lon']+360) % 360)
new_lons
```

```python
ds_cmap2=ds_cmap.assign_coords(lon=new_lons)
ds_cmap2['lon']
```

```python
ds_cmap2=ds_cmap2.sortby(ds_cmap2['lon'])
ds_cmap2
```

```python
precip_check2=ds_cmap2['precip'].sel(lat=0,
                                time=ds_cmap2['time'][0],
                                method='nearest')
plt.plot(ds_cmap2['lon'],precip_check2)
plt.title('CMAP Precip 0-360')
```

```python
ds_cmap3=ds_cmap.assign_coords(lon=((ds_cmap['lon']+360)%360))
ds_cmap3=ds_cmap3.sortby(ds_cmap3['lon'])
```

```python
precip_check3=ds_cmap3['precip'].sel(lat=0,
                                  time=ds_cmap3['time'][0],
                                  method='nearest')
plt.plot(ds_cmap3['lon'],precip_check3)
plt.title('CMAP Precip 0-360 two lines')
```

```python
ds_cmap4=ds_cmap3.assign_coords(lon=(((ds_cmap3['lon']+180)%360)-180))
ds_cmap4=ds_cmap4.sortby(ds_cmap4['lon'])
ds_cmap4
```

```python
precip_check4=ds_cmap4['precip'].sel(lat=0,
                                  time=ds_cmap4['time'][0],
                                  method='nearest')
plt.plot(ds_cmap4['lon'],precip_check4)
plt.title('CMAP Precip -180-180 two lines')
```

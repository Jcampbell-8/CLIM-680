```python
import xarray as xr
import numpy as np
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import matplotlib.colors as colors

from clim_utils import label_latlons
```

# **Dataset Info**
## Observed mean daily precipiation rates averaged throughout a month.
## Grib file format found in the CMAP precipiation database on Cola server.

```python
path='/homes/jcampb8/data-shell/thesis/'
file='enso4_1979-2018.cmap.monthly.precip.mon.mean.nc'
```

```python
ds=xr.open_dataset(path+file)
ds
```

### Using `reindex` to reverse the latitudes and make countour plot

```python
ds=ds.reindex(lat=list(reversed(ds['lat'])))
```

```python
plt.contourf(ds['precip'][0,:,:])
```

### Pick a point (Panama City, Panama) and plot all precip data

```python
ds_pt=ds.sel(lat=9,lon=360-79,method='nearest')
plt.plot(ds_pt['time'],ds_pt['precip'])
plt.title('Panama City Precipitation (mm/day)')
plt.savefig('PCmean.png')
```

### Group data by month with `groupby`

```python
ds_climo=ds.groupby('time.month').mean()
ds_climo
```

### Select out climotological and two different years to plot

```python
ds_ptclimo=ds_climo.sel(lat=9,lon=360-79,method='nearest')
ds_pt1985=ds_pt.sel(time=slice('1985-01-01','1985-12-01'))
ds_pt2015=ds_pt.sel(time=slice('2015-01-01','2015-12-01'))

x_lab=np.arange(1,13,1) ##To add x-label for months to display accurately
#mon_lan=['J','F]etc for month labeling on x-axis
plt.plot(ds_pt1985['precip'])
plt.plot(ds_pt2015['precip'])                    
plt.plot(ds_ptclimo['precip'])
plt.title('Panama City Precipitation (mm/day)')
plt.legend(['1985','2015','Climo'])
plt.savefig('PCclimo.png')
```

### Select out precipitation anomalies

```python
ds_anoms=ds.groupby('time.month')-ds_climo
ds_anoms
```

```python
#ds_anoms.to_netcdf('/homes/jcampb8/anoms_1979-2018.cmap.monthly.precip.mon.mean.nc')
```

### Plot those same anomalies at a point (Panama City, Panama)

```python
ds_anomspt=ds_anoms.sel(lat=9,lon=360-79,method='nearest')
plt.plot(ds_anomspt['time'],ds_anomspt['precip'])
plt.title('Panama City Precipitation Anomalies (mm/day)')
#plt.savefig('PCanoms.png')
```

### Find the about the driest year and plot precip vs. climotology

```python
panama_pt=ds_anomspt['precip']
panama_ptclimo=ds_ptclimo['precip']
panama_pt2016=panama_pt.sel(time=slice('1985-01-01','1985-12-31'))
mon_lab=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.plot(mon_lab,panama_pt2016)
plt.plot(mon_lab,panama_ptclimo)
plt.legend(['1985','Climo'])
plt.title('CMAP Precipitation, Panama City, Panama')
plt.xlabel('Precipitation Anomalies (8N,79W)')
plt.ylabel('mm/day')
plt.savefig('PC1985comparison.png')
```

### Plot precipitation anomalies for climo and two other years with all labels, colormaps, etc.

```python
import matplotlib.colors as colors ##Set zero to middle of color bar

ds_anoms_1980s=ds_anoms.sel(time=slice('1983-01-01','1985-12-01'))

clevs=np.arange(-6,14,2)
fig=plt.figure(figsize=(11,8.5))
ax=plt.axes(projection=ccrs.Robinson())

divnorm=colors.DivergingNorm(vcenter=0) ##Set zero to middle of color bar

cs=ax.contourf(ds_anoms_1980s['lon'],ds_anoms_1980s['lat'],
               ds_anoms_1980s['precip'][0,:,:],clevs,
               transform=ccrs.PlateCarree(),
               cmap='RdBu_r',extends='both',
               norm=divnorm) ##Set zero to middle of color bar

ax.coastlines()
ax.gridlines()
cbar=plt.colorbar(cs,shrink=0.7,orientation='horizontal',
                 label='Precipitation (mm/day)')
plt.title('Precipitation Anomalies, Panama City, 1983-1985')
plt.savefig('PC1980s.png')
```
